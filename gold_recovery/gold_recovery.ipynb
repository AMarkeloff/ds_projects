### Проверка расчета эффективности обогащения

Проверим, что *эффективность обогащения* **рассчитана правильно**. 

Вычислим её на обучающей выборке для признака `rougher.output.recovery`. 

Будем использовать следующую формулу для **эффективности обогащения**:

$$ {\bf Recovery} =\frac{C \times (F - T)}{F \times (C - T)}\times100\%  ,$$

где:

$C$ — доля золота в концентрате после флотации/очистки (`rougher.output.concentrate_au`);

$F$ — доля золота в сырье/концентрате до флотации/очистки (`rougher.input.feed_au`);

$T$ — доля золота в отвальных хвостах после флотации/очистки (`rougher.output.tail_au`).

Для удобства восприятия кода в формуле введём переменные:

```python
#сохраняем train_initial['rougher.output.concentrate_au'] в перменную c_rougher_au
c_rougher_au = train_initial['rougher.output.concentrate_au']

#сохраняем train_initial['rougher.input.feed_au'] в перменную f_rougher_au
f_rougher_au = train_initial['rougher.input.feed_au']

#сохраняем train_initial['rougher.output.tail_au'] в перменную t_rougher_au
t_rougher_au = train_initial['rougher.output.tail_au']
```

Теперь запишем формулу, правильность которой стало проще контролировать после введения переменных:

```python
#вычисляем эффективности обогащения:
recovery_rougher_au = c_rougher_au * (f_rougher_au - t_rougher_au)/(f_rougher_au*(c_rougher_au-t_rougher_au))*100
```

Выведем результат - первые 5 строк:

```python
#выводим первые 5 строк рассчитанной эффективности обогащения
recovery_rougher_au.head()
```

А также искомый столбец:

```python
#выводим первые 5 строк столбца 'rougher.output.recovery' обучающего набора данных
train_initial['rougher.output.recovery'].head()
```

По первым 5 строкам наблюдается полное соответствие.

Найдём **MAE** между расчётами и значением признака, используем встроенную функцию `mean_absolute_error`: 

```python
#находим MAE между расчётами и значением целевым признаком 'rougher.output.recovery' в обучающем наборе данных
mean_absolute_error(train_initial['rougher.output.recovery'],
                    recovery_rougher_au)
```

**Уровень MAE чрезвычайно мал**, что позволяет сделать вывод, что вычисления произведены **верно**, а также, что в исходных данных нет ошибки в вычислении целевого признака.

Для более глубокого анализа визуализируем выборки с помощью графика **"ящик с усами"**. Сделаем это с помощью функции построения графика. На вход функция будет принимать наименование металла (лучше с большой буквы), на выходе - будем получать готовый для анализа график:

```python
# функция постороения графика "ящик с усами" для рассматрвиаемого металла,
# только для интересущих нас этапов очистки
def box_plot_metalls(metall_name):
    
    #размер графика
    plt.figure(figsize=(15,8))
    
    sns.boxplot(data=full_initial[['rougher.input.feed_'+metall_name.lower(),
                                   'rougher.output.concentrate_'+metall_name.lower(),
                                   'primary_cleaner.output.concentrate_'+metall_name.lower(),
                                   'final.output.concentrate_'+metall_name.lower()]])
    
    #название графика
    plt.title('Диаграммы размаха для концентрации '+metall_name+' на разных стадиях', fontsize=16 )
    
    #подписи осей
    plt.xlabel('Стадия технологического процесса', fontsize=16)
    plt.ylabel('Концентрация '+metall_name, fontsize=16)
    
    #задаём размер значений на оси абсцисс и на оси ординат
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=14)
    
    #отображение сетки
    plt.grid()
    plt.show()
```

#### Анализ плотности распределения концентраций металлов от времени на различных этапах

Для исследовательского анализа сделаем ещё одну операцию: построим плотности распределения концентраций для каждого из металлов на разных стадиях с использованием библиотеки `seaborn`.

Для этого создадим специальную функцию, которая будет принимать название металла (можно с большой буквы), а на выходе - строить графики плотности распределения концентраций для каждого из этапов.

Перейдём к записи данной функции:

```python
def dist_plot_metalls(metall_name):
    
    #размер графика
    plt.figure(figsize=(15,8))
    
    for name in ['rougher.input.feed_'+metall_name.lower(),
                 'rougher.output.concentrate_'+metall_name.lower(),
                 'primary_cleaner.output.concentrate_'+metall_name.lower(),
                 'final.output.concentrate_'+metall_name.lower()]:
        sns.distplot(full_initial[name],
                    label=name)
    
    #название графика
    plt.title('Плотность распределения концентрации '+metall_name+' на разных стадиях', fontsize=16 )
    
    #подписи осей
    plt.xlabel('Концентрация', fontsize=16)
    plt.ylabel('Плотность распределения концентрации '+metall_name, fontsize=16)
    
    #задаём размер значений на оси абсцисс и на оси ординат
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=14)
    
    plt.legend(fontsize=14)
    
    #отображение сетки
    plt.grid()
    plt.show()
```

Функция готова, построим первый график для золота:

```python
dist_plot_metalls('Au')
```

Результат аналогичен предыдущим шагам исследования: концентрация сначала увеличивается, затем снижается. Также как и для золота - есть значения с нулевой концентрацией.

Плотность распределения концентрации **свинца**:

```python
dist_plot_metalls('Pb')
```

Концентрация свинца сначала немного снижается, затем мало меняется. Есть около нулевые значения, выраженные небольшими пиками в области нуля. 

Таким образом, анализ плотностей распределения концентраций **подтверждает ранее сделанные выводы**.

#### Статистическая проверка распределений размера гранул сырья

Для проверки гипотезы сначала сформулируем нулевую гипотезу $H_0$:**средние значения распределений размера гранул в обучающем и тестовом наборах данных равны (не отличаются)**.

Тогда двусторонняя альтернативная гипотеза $H_1$: **средние значения распределений размера гранул в обучающем и тестовом наборах данных не равны**, они могут отличаться, как в большую так и в меньшую сторону.

Чтобы проверить гипотезу о равенстве среднего двух генеральных совокупностей **по взятым из них выборкам**, применим метод `scipy.stats.ttest_ind (array1, array2, equal_var)`.

Выберем критический уровень значимости $\alpha$ равным **5%**.

```python
alpha = 0.05  # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
```

Создадим функцию для проверки статистических гипотез, будем использовать t-тест `ttest_ind` из библиотеки `scipy`: 

```python
#функция проверки статистических гипотез
def statistical_check(sample_1, sample_2, alpha):
    
    #использование ttest_ind для двух выборок
    results = st.ttest_ind(sample_1, sample_2)
    
    #вывод p-value
    print('p-значение:', results.pvalue)
    
    #проверка правильности нулевой гипотезы
    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не получилось отвергнуть нулевую гипотезу') 
```

Теперь используем эту функцию для проверки гипотез:

```python
#используем функцию с уровнем статистической значимости alpha
statistical_check(feed_size['test_primary_cleaner.input.feed_size'], 
                  feed_size['train_primary_cleaner.input.feed_size'],
                  alpha)
```

```python
#используем функцию с уровнем статистической значимости alpha
statistical_check(feed_size['test_rougher.input.feed_size'], 
                  feed_size['train_rougher.input.feed_size'],
                  alpha)
```

Получается, что по результатам статистического теста мы должны **отвергнуть нулевую гипотезу** и сделать вывод о неравенстве средних значений рассматриваемых выборок. 

Но так как нам **не нужна высокая точность**, распределения **визуально похожи**, то не будем удалять этот признак из рассмотрения.

### Функция для вычисления итоговой sMAPE

Для решения задачи введём новую метрику качества — **sMAPE** (англ. *Symmetric Mean Absolute Percentage Error*, «симметричное среднее абсолютное процентное отклонение»).

Она похожа на **MAE**, но выражается не в абсолютных величинах, а в относительных. Почему симметричная? Она одинаково учитывает масштаб и целевого признака, и предсказания.

Метрика **sMAPE** вычисляется так:

$$ {\bf sMAPE} =  \frac{1}{\rm N} \sum_{i=1}^{{\rm N}} \frac{|y_{i}-\widehat{y_{i}}|} {(|y_{i}|+|\widehat{y_{i}}|) \over 2} \times100\%, $$

где 
- $y_{i}$ - значение целевого признака для объекта с порядковым номером $i$ в выборке, на которой измеряется качество,
- $\widehat{y_{i}}$ - значение предсказания для объекта с порядковым номером $i$, например, в тестовой выборке,
- $\rm N$ - количество объектов в выборке,
- $\sum_{i=1}^{{\rm N}}$ - суммирование по всем объектам выборки ($i$ меняется от $1$ до $\rm N$).



В задаче нужно спрогнозировать сразу две величины:
- эффективность обогащения чернового концентрата `rougher.output.recovery`;
- эффективность обогащения финального концентрата `final.output.recovery`.

Поэтому итоговая метрика складывается из двух величин:

$$ Итоговое\ {\bf sMAPE} = 25\% \times {\rm sMAPE(rougher)} + 75\% \times {\rm sMAPE(final) }.$$

Напишем функцию для вычисления итоговой sMAPE.

```python
def final_sMAPE(targets, predictions):
    
    sMAPE_rougher = ((abs(targets['rougher.output.recovery'] - predictions['rougher.output.recovery'])/
                      ((abs(targets['rougher.output.recovery'])+abs(predictions['rougher.output.recovery']))/2)*100)
                     .sum()/len(targets['rougher.output.recovery']))
    
    sMAPE_final = ((abs(targets['final.output.recovery'] - predictions['final.output.recovery'])/
                    ((abs(targets['final.output.recovery'])+abs(predictions['final.output.recovery']))/2)*100)
                   .sum()/len(targets['final.output.recovery']))
    
    final_SMAPE = (0.25*sMAPE_rougher+0.75*sMAPE_final)
    
    return final_SMAPE
```

Далее будем использовать функцию для вычисления метрики качества во время кросс-валидации.

2. На каждом шаге цикла будем разбивать переменные `features_train` и `target_train` на выборки `features_train_cv`, `target_train_cv`, `features_valid_cv` и `target_valid_cv` так, чтобы в них были только наблюдения с нужными номерами.

Зададим максимальную глубину решающего дерева:

```python
max_forest_depth = 6
```

Будем варьировать количество деревьев в диапазоне:

```python
min_estimators = 10 #нижняя граница
max_estimators = 70 #верхняя граница
```

С шагом:

```python
step_estimators = 10 #шаг по количеству деревьев
```

Теперь с помощью трёх вложенных циклов:
- будем варьировать количество деревьев,
- будем варьировать глубину решающего дерева,
- для каждой комбинации глубина и количество деревьев будем выполнять кросс-валидацию и оценивать среднее на 5 блоках значение sMAPE.

```python
avarage_scores_rfr = [] #здесь сохраним средние результаты sMAPE после кросс-валидации
array_depth_rfr = []    #здесь будем сохранять значения глубины решающего дерева
array_est_rfr = []      #здесь сохраним количество деревьев для каждого шага цикла

#счетчики, которые обновим в процессе работы цикла
best_forest_score = 100
best_result_forest_depth = 0
best_result_est = 0

# зададим размер блока для кросс-валидации
sample_size = int(len(features_train)/5)

#цикл по количеству деревьев
for est in range(min_estimators,max_estimators,step_estimators):
    
    #цикл по значениям глубины решающего дерева
    for depth in range(1,max_forest_depth+1):
        
        #цикл для кросс-валидации
        for i in range(0, len(features_train), sample_size):
            
            #здесь сохраним sMAPE для каждого шага цикла кросс-валидации
            scores_rfr = []
            
            # запишем массив из индексов для валидационного блока    
            valid_indexes = list(range(i, i+sample_size))   
            train_indexes = list(range(0,i))+list(range(i+sample_size,len(features_train)))
                
            # запишем массив из индексов для обучающей выборки
            #if i==0:
            #    train_indexes = list(range(i+sample_size,len(features_train)))
            #    
            #elif i==len(features_train)-sample_size:
            #    train_indexes = list(range(0,i))
            #    
            #else:
            #    train_indexes = list(range(0,i))+list(range(i+sample_size,len(features_train)))
                
            
            #разобъём переменные features_train и target_train на выборки 
            #features_train_cv, target_train_cv, features_valid_cv, target_valid_cv
            features_train_cv = features_train[features_train.index.isin(train_indexes)]
            features_valid_cv = features_train[features_train.index.isin(valid_indexes)]
            
            target_train_cv = target_train[target_train.index.isin(train_indexes)].reset_index(drop=True)
            target_valid_cv = target_train[target_train.index.isin(valid_indexes)].reset_index(drop=True)
        
            #введём модель случайного леса
            model_rfr = RandomForestRegressor(random_state=12345, 
                                              max_depth=depth,
                                              n_estimators=est)
            
            #обучим модель
            model_rfr = model_rfr.fit(features_train_cv, target_train_cv)
            
            #находим предсказания
            predictions_valid_cv = model_rfr.predict(features_valid_cv)
        
            #преобразуем предсказания в датафрейм
            predictions_valid_cv = pd.DataFrame(predictions_valid_cv, columns =['rougher.output.recovery', 
                                                                                'final.output.recovery']) 
            #находим метрику качества
            score = final_sMAPE(target_valid_cv, predictions_valid_cv)
            
            #добавляем её в массив
            scores_rfr.append(score)
    
        #итоговая метрика - среднее значение sMAPE по результатам кросс-валидации
        final_score_rfr = sum(scores_rfr)/len(scores_rfr)
    
        #сохраняем лучшую модель
        if final_score_rfr < best_forest_score:
            best_forest_score = final_score_rfr
            best_model_forest_depth = model_rfr
            best_result_est = est
            best_result_forest_depth = depth
    
        #сохраняем данные для построения графиков
        avarage_scores_rfr.append(final_score_rfr)
        array_depth_rfr.append(depth)
        array_est_rfr.append(est)
        
print('best_forest_score (sMAPE)', best_forest_score)
print('best_result_est', best_result_est)
print('best_result_forest_depth', best_result_forest_depth)
```

##### Кросс-валидация с помощью `cross_val_score`

Повторим ранее выполненные для решающего дерева шаги: кросс-валидаци с помощью `cross_val_score`, итоговая таблица и график.

```python
avarage_scores_rfr_1 = [] #здесь сохраним средние результаты sMAPE после кросс-валидации
array_depth_rfr_1 = []    #здесь будем сохранять значения глубины решающего дерева
array_est_rfr_1 = []      #здесь сохраним количество деревьев для каждого шага цикла

#счетчики, которые обновим в процессе работы цикла
best_forest_score_1 = 100
best_result_forest_depth_1 = 0
best_result_est_1 = 0

#цикл по количеству деревьев
for est in range(min_estimators,max_estimators,step_estimators):
    
    #цикл по значениям глубины решающего дерева
    for depth in range(1,max_forest_depth+1):
        
        #введём модель случайного леса
        model_rfr_1 = RandomForestRegressor(random_state=12345, 
                                            max_depth=depth,
                                            n_estimators=est)
        
        scores_rfr_1 = cross_val_score(model_rfr_1, 
                                       features_train, 
                                       target_train,
                                       scoring=cvs_score_sMAPE,
                                       cv=5)
        
        
        
        final_score_rfr_1 = sum(scores_rfr_1)/len(scores_rfr_1)
        
        #сохраняем лучшую модель
        if abs(final_score_rfr_1) < best_forest_score_1:
            best_forest_score_1 = abs(final_score_rfr_1)
            best_result_est_1 = est
            best_result_forest_depth_1 = depth
        
        avarage_scores_rfr_1.append(abs(final_score_rfr_1))
        array_depth_rfr_1 .append(depth)
        array_est_rfr_1.append(est)
        
print('best_forest_score (sMAPE)', best_forest_score_1)
print('best_result_est', best_result_est_1)
print('best_result_forest_depth', best_result_forest_depth_1)
```

```python
#датафрейм с результатами исследования модели случайного леса с учетом глубины решающего дерева
result_forest_depth_1 = pd.DataFrame({'n_estimators':np.array(array_est_rfr),
                                      'depth':np.array(array_depth_rfr_1),
                                      'forest_score':np.array(avarage_scores_rfr_1)})
```
