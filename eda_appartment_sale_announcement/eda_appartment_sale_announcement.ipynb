# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Цель исследования**

Установить параметры для автоматизированного определения рыночной стоимости объектов недвижимости.

Показатель рыночной стоимости объекта недвижимости - **стоимость квадратного метра**. 
Нам предстоит определить параметры, оказывающие наибольше влияние на стоимость квадратного метра и проследить зависимость от них.

Получаем **разные этажи** и **отсутствие пропусков** (об отсутсвие пропусков знаем из самого первого вызова функции `info()` ).

**Возникает идея** заменить пропуски в этажности здания на соответствующий указанный этаж. Но тогда мы **увеличим выборку** в категории: **последний этаж**. 

Посмотрим сможем ли мы в дальнейшем провести исследование влияния последнего этажа на цену квадратного метра. Сколько всего у нас данных **по последнему этажу** без учета пропущенных значений во `floors_total`, какой процент от общего набора данных?

```python
#последние, не 1ые этажи в исходной выборке, их уникальные значения и соответствующее им количество
df[(df['floors_total']==df['floor']) & (df['floor']!=1)]['floors_total'].value_counts()
```

```python
#общее количество последних этажей, не 1ых
df[(df['floors_total']==df['floor']) & (df['floor']!=1)]['floors_total'].count()
```

```python
#процент объявлений с последним этажом от общего количества объявлений
(
df[(df['floors_total']==df['floor']) & (df['floor']!=1)]['floors_total'].count()
    /
len(df['floor'])
    *
    100
)
```

Больше **3000** объектов недвижимости, в которых предлагается жильё на последнем этаже. 

Это **14%** от исходного набора данных - достаточно хорошая выборка.

Посмотрим насколько мы исказим данные, если будем считать пропуски в `floors_total` равными `floor`, то есть последними этажами, тогда мы добавим примерно:

```python
#процент объявлений с пропущенными значениями в столбце 'floors_total'
#от общего количества объявлений с последним этажом

(
    df['floors_total'].isna().sum()
    /
    df[(df['floors_total']==df['floor']) & (df['floor']!=1)]['floors_total'].count()
    *
    100
)
```

Почти **2,6%** значений мы искуственно введём в категорию **последний этаж**. Это может исказить результаты исследования в категории **последний этаж**.

```python
# процент объявлений в категории этажей  "не последний, не первый (другой)"
# от общего количества объявлений

(
df[(df['floors_total']!=df['floor']) & (df['floor']!=1)]['floors_total'].count()
    /
len(df['floor'])
    *100
)
```

Если мы заполним пропуски во `floors_total` так, чтобы эти объявления **не стали** *жильём на последнем этаже*, то мы не потеряем данные и не исказим результаты, так как добавим незначительный процент к выборке по данной категории:

```python
# процент объявлений с пропущенными значениями в столбце 'floors_total'
# от общего количества объявлений в категории этажей  "не последний, не первый (другой)"
(
df['floors_total'].isna().sum()
    /
df[(df['floors_total']!=df['floor']) & (df['floor']!=1)]['floors_total'].count()
    *
    100
)
```

Постепенно мы приходим к решению:

- **Нам предстоит** выделить  3 категории квартир по этажам:  **первый**, **последний** и **другой**.
- Соответственно столбец `floors_total` поможет нам найти **последние этажи**, только в этом его полезная функция.
- Увеличивать выборку категории **последний этаж** мы не будем, чтобы не исказить результаты.

Между тремя столбцами `total_area`, `kitchen_area`, `living_area` есть понятная **связь:** **сумма** *площади кухни* и *жилой площади* **не должна превышать** *общую площадь*, или

$$S_{общая}>(S_{жилая}+S_{кухни})$$

Если в каких-то объявлениях обнаружен этот факт, то это будет свидетельствовать об ошибке введённых продавцом данных.

Перед тем, как заменить пропущенные значения в стобцах `kitchen_area` и `living_area` посмотрим для какого количества объявлений не выполняется данное условие, проверим выполнение условия

$$S_{общая}-S_{жилая}-S_{кухни}>0$$

```python
#проверим исходные объявления: все ли продавци верно указали площади
#кухни, жилой и общей?
((df['total_area']-df['kitchen_area']-df['living_area'])<0).sum()
```

В 65 объявлениях $S_{общая}<(S_{жилая}+S_{кухни})$, в 65 объявлениях продавцы **допустили ошибку**. Будем иметь это в виду.

### Категория по этажу - первый, последний, другой

С помощью данных в столбцах `floor` и `floors_total` запишем функцию, которая выполнит категоризацию данных по этажу: первый, последний или другой.

Принцип категоризации:
- **первый этаж** - строки со значением 1 в столбце `floor`,
- **последний этаж** - строки, в которых значение `floor` = `floors_total` и не равно 1 (не равно единице, чтобы не дублировать 1ые этажи для одноэтажных зданий),
- **другой** - все прочие объекты недвижимости, не удовлетворяющие первым двум условиям.

Теперь запишем это в виде функции:

```python
#создаём функцию floor_category
def floor_category(column1, column2):
    
    #проверяем относится ли объявление к первому этажу
    if (column1 == 1):
        return 'первый'
    
    #проверяем последний это этаж или нет
    if ((column1 == column2) &
        (column1 != 1)):
        return 'последний'
    
    #если первые два условия не выполнены, то категория - другой
    else:
        return 'другой'
```

Функцию необходимо применить к **двум столбцам сразу** (`floor` и `floors_total`), чтобы получить новый столбец `floor_category`, поэтому для записи кода в одну строчку воспользуемся `lambda`-функцией и методом `apply()`:

```python
#создаём новый столбец 'floor_category', в который добавляем категорию объявления по этажу
# с помощью функции floor_category, lambda-функции и метода apply();
#для категоризации используем данные в столбцах 'floor' и 'floors_total'

df['floor_category'] = df.apply(lambda x: floor_category(x.floor, x.floors_total), axis=1)
```

Оценим интервал значений для столбца `total_area`, которые будут более вероятны исходя из параметров выборки. Нижнее значение у нас ограничено 12 квадратными метрами, а верхнее рассчитаем, зная первый и третий квартили, умножив их разность на 1,5:

Сделаем срез данных от минимального значения выборки `total_area` до значения $Q3+1.5(Q3-Q1)$, где $(Q3-Q1)$ -это $IQR$ (интерквартильный размах).

Посмотрим, как выглядит гистограмма среза данных, она уже чем-то напоминает скорее **распределение Пуассона**, чем нормальное. 

```python
#строим гистограмму по срезу дынных в 'total_area'
plt.figure(figsize=(10,5))
(
    df
    .query('total_area <= (@total_area_Q3 + 1.5 * (@total_area_Q3-@total_area_Q1))')['total_area']
    .hist(bins=25)
)

plt.title('Гистограмма значений общей площади \n по срезу данных')
plt.xlabel('total_area')
plt.show()
```

Из графика также видно, что очень мало объявлений с площадью менее 20 квадратных метров. Их можно будет исключить из рассмотрения.

Теперь мы знаем каким значениям `total_area` можно доверять: от, примерно, 20 квадратных метров, до **114** квадратных метров (по $IQR$ и до 150-200 метров квадратных, исходя из здравого смысла). Будем двигаться дальше.

Посмотрим чему равно $IQR$:

Максимальная цена квартиры составила **11,9 млн руб**, большинство значений выборки (75%) **менее 6 млн. руб.**. Средняя цена 4,85 млн. руб, медиана - 4,4 млн. руб..

Но будем помнить, что если ограничим минимум **3 млн. руб.**, то получем график очень похожий на распределение Пуассона:

```python
#строим гистограмму по срезу с минимальным значением
(
    df
    .query('(3000000 < last_price <= (@last_price_Q3 + 1.5 * (@last_price_Q3-@last_price_Q1)))')['last_price']
    .hist(bins=50)
)
plt.title('Гистограмма цен квартир \n по срезу данных')
plt.xlabel('last_price')
plt.show()
```

```python
#строим гистограмму по срезу 0 < days_exposition < (Q3+1.5IQR)
(
    df.
    query('0 < days_exposition < (@days_exposition_Q3+1.5*@days_exposition_IQR)')
    ['days_exposition']
    .hist(bins=512,figsize=(12,6))
)

plt.title('Гистограмма days_exposition по срезу данных')
plt.xlabel('days_exposition')
plt.show()
```

Очевидно, что есть несколько (примерно 5 шт) **выдающихся значений** в диапазоне 0-100 суток. 

Приблизим, чтобы посмотреть, где они находятся:

```python
#строим гистограмму по срезу 0 < days_exposition < (Q3+1.5IQR)
# в диапазоне 0-100 дней
(
    df.
    query('0 < days_exposition < (@days_exposition_Q3+1.5*@days_exposition_IQR)')
    ['days_exposition']
    .hist(bins=100,range=(0,100),figsize=(15,6))
)

plt.title('Гистограмма days_exposition по срезу данных')
plt.xlabel('days_exposition')
plt.show()
```

Посмотрим какие по счету дни с момента размещения объявления имеют наибольшее количество снятий объявления с продажи:

```python
#смотрим уникальные значения в срезе 0 < days_exposition < 100
#количество каждого из уникальных значений

days_exposition_uniq_0_100 = (df.
                              query('0 < days_exposition < 100')
                              ['days_exposition']
                              .value_counts()
                             )

days_exposition_uniq_0_100.head(10)
```

Очень похоже на **окончание определённых периодов размещения** объявления (платных/бесплатных). 

Поэтому **большое количество снятий объявлений** с продажи в эти дни **не означает**, что они были проданы.

Посмотрим характеристику выборки уникальных значений среза `days_exposition` в диапазоне 0-100 дней.

#### Зависимость от числа комнат

Как зависит цена от числа комнат?

Начнём со стоимости квадратного метра. Количество комнат ограничим - от 1 до 6 (как было определено ранее).
Диаграмма рассеяния будет малоинформативной в данном случае, поэтому воспользуемся `pivot_table`. Запишем в новую переменную `rooms_id_sqm` сводную таблицу для `rooms` и `square_meter_price`, посчитаем выборку для каждой из комнат, оценим среднюю стоимость квадратного метра.

```python
#сводная таблица, показывающая объём выборки для каждой из комнат,
#а также среднюю стоимость квадратного метра
rooms_id_sqm = (
    df
    .query('(square_meter_price<200000) and (0<rooms<=6)')
    .pivot_table(index = ['rooms'], 
                 values='square_meter_price', 
                 aggfunc=(['count','mean']))
)
#избавимся от двойного названия столбцов
rooms_id_sqm.columns=['count','mean_sq_meter_price']

#выведем на экран сводную таблицу
rooms_id_sqm
```

Выборка достаточна для комнат 1-4 комнат. Для 5-6 комнат выборка мала. Построим график зависимости стоимости квадратного метра от количества комнат:

```python
#строим график зависимости стоимости квадратного метра от количества комнат
rooms_id_sqm.plot(y='mean_sq_meter_price', 
                  style='-o',
                  ylim=0,
                  grid=True,
                  figsize=(12,6))

plt.title('Зависимость стоимости квадратного метра от количества комнат')
plt.show()
```

Получаем интересный результат. Стоимость квадратного метра снижается с увеличением количества комнат от 1 до 3: **от 100 тыс. до 90 тыс. руб. за квадратный метр**. На 4 комнатах средняя цена снова немного увеличивается и становится равной 95 тыс. руб. за кв. м.

Посмотрим, как на цену квартиры влияет количество комнат. Построим сводную таблицу и выведем график:

```python
#сводная таблица, показывающая объём выборки для каждой из комнат,
#а также среднюю стоимость объекта недвижимости
rooms_id_lastprice = (
    df
    .query('(0<rooms<=6) & (last_price<12000000)')
    .pivot_table(index = ['rooms'], 
                 values='last_price', 
                 aggfunc=(['count','mean']))
)

rooms_id_lastprice.columns=['count','mean_last_price']

rooms_id_lastprice
```

```python
#строим график зависимости средней стоимости объекта недвижимости от количества комнат
rooms_id_lastprice.plot(y='mean_last_price', 
                        style='-o',
                        ylim=0,
                        grid=True, 
                        figsize=(10,5))

plt.title('Зависимость средней стоимости объекта недвижимости от количества комнат', pad=15)
plt.show()
```

Зависимость стоимости объекта недвижимости явная - чем больше комнат, тем выше стоимость.

##### Этаж

Создадим функцию, которая будет строить график и выводить сводную таблицу для категорийных строковых переменных.

```python
#функция category_group_barh входные данные для которой - тип цены и колонка с категорийной переменной
def category_group_barh (price_type, colum1):
    id_price = ( df_good_sntp_center
    .pivot_table(index = [colum1], 
                 values= price_type, 
                 aggfunc=(['count','mean']))
             )
    #избавимся от двойного названия столбцов
    id_price.columns=['count',price_type+'_mean']
    
    #выведем на экран сводную таблицу
    display(id_price)
    
    #строим столбчутую диаграмму
    (id_price
     .plot(y=price_type+'_mean',
          figsize=(10,6),
          grid=True,
          ylim=0,
          kind='barh'))
```

Применем функцию к столбцу `floor_category`:

```python
#применяем функцию к столбцу floor_category
category_group_barh('square_meter_price','floor_category')
plt.title('Зависимость стоимости квадратного метра от этажа', pad=15)
plt.show()
```

Средняя цена квадратного метра на первом этаже (112,8 тыс. руб.) ниже, чем на последнем (121,3 тыс. руб.), а последний этаж дешевле других этажей (129 тыс. руб.).

Повторим применение функции для `floor_category`, но для цены `last_price`:

```python
#применяем функцию к столбцу floor_category
category_group_barh('last_price','floor_category')
plt.title('Зависимость стоимости объекта недвижимости от этажа', pad=15)
plt.show()
```

В центре Санкт-Петербурга объекты недвижимости на первом этаже значительно дешевле последнего и других этажей, в среднем примерно на 35-40%. Это объяснимо, на первом этаже сложнее избежать городской суеты и шума автотрансопорта.

##### Расстояние до центра

Теперь **запишем функцию** для создания **сводной таблицы** и построения графика **с линией и точками**. Функция будет выдавать результат сразу по двум типам цены - квадратный метр и общая стоимость.

Пригодится для таких категорийных переменных, которые в тоже время являются числами: `rooms`, `floor_category` и `city_centers_nearest_km` и других. 

```python
#функция category_group_plot, входные данные для которой -  колонка с категорийной переменной

#создаём функцию
def category_group_plot (colum1):
    
    #массив типов цены
    array_price = ['square_meter_price', 'last_price']
    
    #перебираем типы цены
    for price_type in array_price:
        
        #делаем сводную таблицу
        id_price = ( df_good_sntp_center
                    .pivot_table(index = [colum1], 
                                 values= price_type, 
                                 aggfunc=(['count','mean']))
                   )
        
        #избавимся от двойного названия столбцов
        id_price.columns=['count',price_type+'_mean']
        
        #выведем на экран сводную таблицу
        display(id_price)
        
        #строим график    
        id_price.plot(y=price_type+'_mean',
                      figsize=(10,5),
                      style='-o',
                      ylim=0,
                      grid=True)
        plt.title('Зависимость '+price_type+'_mean'+' от '+ colum1, pad=15)
        plt.show()
```

Далее применяем функцию для столбца `city_centers_nearest_km`. 
Функция позволяет сделать это сразу для двух типов цены (общая и за кв. метр):

```python
#применяем функцию к столбцу city_centers_nearest_km - расстояние до центра в километрах
category_group_plot('city_centers_nearest_km')
```

Делаем вывод, что цена **квадратного метра** в центре Санкт-Петербурга слабо зависит от удалённости от центра. В свою очередь, общая цена объекта недвижимости снижается по мере удаления от центра.
